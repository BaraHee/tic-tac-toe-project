```python
from IPython.display import clear_output #step 1
def display_board(board):
    clear_output()
    print(board[1] + '|' + board[2] + '|' + board[3])
    print(board[4] + '|' + board[5] + '|' + board[6])
    print(board[7] + '|' + board[8] + '|' + board[9])
```

```python
def players_input(): #step2
    player1=''
    player2=''
    
    while player1 not in ['X','O']:
        
        player1=input("Please Pick X or O to start the game!").upper() #upper so it counts x and o
        
        if player1 == 'X':
            player2 = 'O'
        elif player1 == 'O':
            player2 = 'X'
            
    print(f'\nPlayer1 will be {player1}! and Player2 will be {player2}!')
    return player1,player2
```

```python
from random import randint #step 5

def choose_first():
    if randint(1,2) == 1:
        return 'player1'
        
    else:
        return 'player2'
```

```python
def check_space(board,position): #step 6
    
    if board[position]== 'X' or board[position]== 'O':
        return False #already marked
    else:
        return True #empty space
```

```python
def full_board_check(board): #step7
    for i in range(1,10):
        if board[i] == ' ':
            return False #false theres still space
    return True #true is full 
```

```python
def player_choice(board,check_space): #step 8
    position = 0
    while position not in range(1,10) or not check_space(board,position):
        position=int(input('pick ur next positon from 1-9: '))
    return position 
```

```python
def replay():
    answer=''
    while answer not in ['Yes','No']:
        answer=input('Would you like to contiue playing?: Yes or No')
        if answer == 'Yes':
            clear_output()
            theBoard=[' ']*10
        else:
            break
    return answer
```

```python
while True:
    theBoard = [' '] * 10
    player1_marker, player2_marker = players_input()
    turn = choose_first()
    print(f"{turn} plays first!")

    play_game = input('Are you ready to play? Enter Yes or No: ').lower()
    if play_game.startswith('y'):
        game_on = True
    else:
        game_on = False

    while game_on:
        if turn == 'Player 1':
            # Player 1 turn
            display_board(theBoard)
            position = player_choice(theBoard)
            place_marker(player1_marker, position, theBoard)

            if win_check(player1_marker, theBoard):
                display_board(theBoard)
                print('Congratulations! Player 1 has won!')
                game_on = False
            else:
                if full_board_check(theBoard):
                    display_board(theBoard)
                    print("It's a draw!")
                    break
                else:
                    turn = 'Player 2'
        else:
            # Player 2 turn
            display_board(theBoard)
            position = player_choice(theBoard)
            place_marker(player2_marker, position, theBoard)

            if win_check(player2_marker, theBoard):
                display_board(theBoard)
                print('Congratulations! Player 2 has won!')
                game_on = False
            else:
                if full_board_check(theBoard):
                    display_board(theBoard)
                    print("It's a draw!")
                    break
                else:
                    turn = 'Player 1'

    if not replay():
        break
```
