```python
from IPython.display import clear_output
from random import randint

def display_board(board):
    clear_output()
    print(board[1] + '|' + board[2] + '|' + board[3])
    print(board[4] + '|' + board[5] + '|' + board[6])
    print(board[7] + '|' + board[8] + '|' + board[9])
```

```python
def players_input():
    player1 = ''
    player2 = ''
    while player1 not in ['X','O']:
        player1 = input("Please Pick X or O to start the game! ").upper()
        if player1 == 'X':
            player2 = 'O'
        elif player1 == 'O':
            player2 = 'X'
    print(f'\nPlayer1 will be {player1}! and Player2 will be {player2}!')
    return player1, player2
```

```python
def place_marker(marker, position, board):
    board[position] = marker
```

```python
def win_check(marker, board):
    return ((board[1] == board[2] == board[3] == marker) or
            (board[4] == board[5] == board[6] == marker) or
            (board[7] == board[8] == board[9] == marker) or
            (board[1] == board[4] == board[7] == marker) or
            (board[2] == board[5] == board[8] == marker) or
            (board[3] == board[6] == board[9] == marker) or
            (board[1] == board[5] == board[9] == marker) or
            (board[3] == board[5] == board[7] == marker))
```

```python
def choose_first():
    if randint(1,2) == 1:
        return 'Player 1'
    else:
        return 'Player 2'
```

```python
def check_space(board, position):
    return board[position] == ' '
```

```python
def full_board_check(board):
    for i in range(1, 10):
        if board[i] == ' ':
            return False
    return True
```

```python
def player_choice(board):
    position = 0
    while True:
        try:
            position = int(input('Pick your next move (1-9): '))
        except ValueError:
            print("Please enter a number!")
            continue
        if position not in range(1, 10):
            print("Invalid position! Choose 1-9.")
        elif not check_space(board, position):
            print("That position is already marked!")
        else:
            break
    return position
```

```python
def replay():
    answer = ''
    while answer not in ['Yes', 'No']:
        answer = input('Would you like to continue playing? Yes or No: ')
        if answer not in ['Yes','No']:
            print("Please enter Yes or No.")
    return answer == 'Yes'
```

```python
while True:
    theBoard = [' '] * 10
    player1_marker, player2_marker = players_input()
    turn = choose_first()
    print(f"{turn} plays first!")

    play_game = input('Are you ready to play? Enter Yes or No: ').lower()
    if play_game.startswith('y'):
        game_on = True
    else:
        game_on = False

    while game_on:
        if turn == 'Player 1':
            # Player 1 turn
            display_board(theBoard)
            position = player_choice(theBoard)
            place_marker(player1_marker, position, theBoard)

            if win_check(player1_marker, theBoard):
                display_board(theBoard)
                print('Congratulations! Player 1 has won!')
                game_on = False
            else:
                if full_board_check(theBoard):
                    display_board(theBoard)
                    print("It's a draw!")
                    break
                else:
                    turn = 'Player 2'
        else:
            # Player 2 turn
            display_board(theBoard)
            position = player_choice(theBoard)
            place_marker(player2_marker, position, theBoard)

            if win_check(player2_marker, theBoard):
                display_board(theBoard)
                print('Congratulations! Player 2 has won!')
                game_on = False
            else:
                if full_board_check(theBoard):
                    display_board(theBoard)
                    print("It's a draw!")
                    break
                else:
                    turn = 'Player 1'

    if not replay():
        break
```
