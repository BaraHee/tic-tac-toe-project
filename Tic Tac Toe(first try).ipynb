```python
from IPython.display import clear_output
def display_board(board):
    clear_output()
    print(board[1] + '|' + board[2] + '|' + board[3])
    print(board[4] + '|' + board[5] + '|' + board[6])
    print(board[7] + '|' + board[8] + '|' + board[9])
```

```python
test_board = ['#','X','O','X','O','X','O','X','O','X']
display_board(test_board)
```

```python
def player_input(): #make it ignores lowercase letters later
    player1=''
    player2=''
    alist=['X','O']
    
    player1=input('Player1 Pick X or O') #this line of code is absolutley unneccessary i did it for fun
    if player1 == 'X':
            player2 ='O'
            
    elif player1 =='O':
            player2 ='X'

    while player1 not in alist:
        player1=input('Wrong input,Please Pick X or O')
        
        if player1 == 'X':
            player2 ='O'
            
        elif player1 =='O':
            player2 ='X'
    print (f'Player1 is {player1} and Player2 is {player2}')
    return player1,player2
```

```python
player1,player2=player_input()
```

```python
def mark_position(player1,player2,board):
    
    choice1=int(input('Player1 Pick your position: ')) #round 1
    
    board[choice1]=player1

    print(display_board(board))
    
#=======================================================================================
    choice2=int(input('Player2 Pick your position: ')) #round 2
    
    board[choice2]=player2

    print(display_board(board))
    
#=======================================================================================

    choice3=int(input('Player1 Pick your position: ')) #round 3
    
    board[choice3]=player1

    print(display_board(board))
    
#=======================================================================================
    
    choice4=int(input('Player2 Pick your position: ')) #round 4
    
    board[choice4]=player2

    print(display_board(board))
    
#=======================================================================================

    choice5=int(input('Player1 Pick your position: ')) #round 5
    
    board[choice5]=player1

    print(display_board(board))
    
#=======================================================================================

    choice6=int(input('Player2 Pick your position: ')) #round 6
    
    board[choice6]=player2

    print(display_board(board))
    
#=======================================================================================

    choice7=int(input('Player1 Pick your position: ')) #round 7
    
    board[choice7]=player1

    print(display_board(board))
    
#=======================================================================================
    
    choice8=int(input('Player2 Pick your position: ')) #round 8
    
    board[choice8]=player2

    print(display_board(board))
    
#=======================================================================================

    choice9=int(input('Player1 Pick your position: ')) #round 9
    
    board[choice9]=player1

    print(display_board(board))
    
    return display_board(board)
```
